```python
from dbhelper import PgDB, pd_query
import datetime
```

```python
query = '''
        select 
            *
        from weather;
        '''

df = pd_query(query)
```

```python
dfmedian = df.groupby([df.date.dt.month, df.date.dt.day]).median().drop('predicted', 1)
```

```python
dfmedian['date'] = [str(datetime.date(2016, a, b)) for a, b in dfmedian.index]
```

```python
dfmedian.head()
```

```python
dfmedian.rain = dfmedian.rain.astype(bool)
dfmedian.snow = dfmedian.snow.astype(bool)
dfmedian.fog = dfmedian.fog.astype(bool)
dfmedian.thunderstorm = dfmedian.thunderstorm.astype(bool)
```

```python
ins_query = "insert into weather values('%s', %s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s, %s)"
with PgDB() as db:
    for row in dfmedian.itertuples():
        date = row[-1]
        print ins_query % tuple([date] + list(row[1:-1]) + [True])
        db.cur.execute(ins_query % tuple([date] + list(row[1:-1]) + [True]))
#         break

    db.conn.commit()
```
